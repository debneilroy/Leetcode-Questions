## Approach : Sliding Window

<img src = "PIS2.png">

```python
def checkInclusion(s1, s2):
    
    # working with list values [0, 1, 2,...25] instead of 'a' - 'z'
    A = [ord(x) - ord('a') for x in s1] 
    B = [ord(x) - ord('a') for x in s2]
    print('A is', A)
    print('B is', B)
    
    target = [0] * 26
    for x in A:
        target[x] += 1
        
    print('Target is', target)
    print('----------------------------------------------------------------------------------------')
        
    window = [0] * 26
    for i, x in enumerate(B):
        print('Index, value', i, x)
        window[x] += 1
        print(window)
        if i >= len(A):
            window[B[i-len(A)]] -= 1
            print(window)
        if window == target:
            return True
        print('------------------------------------------------------------------------------')
                   
    return False
```

```python
checkInclusion("ab", "eidbaooo")
```
